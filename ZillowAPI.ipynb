```python
def set_url_parameter(url, name, value):
    """Take an input URL template and key-value string, one pair each call"""

    uri = urlparse.urlparse(url)
    
    #convert pr.query from string to dict    
    query_dict = urlparse.parse_qs(uri.query)#Parse a query string given as a string argument
    query_dict[name]= value

    #e.g. http://stackoverflow.com/a/10233141/7782 
    #convert a mapping object or a sequence (doseq=True) of two-element tuples to a “percent-encoded” string
    #suitable to pass to urlopen() as the optional data argument.
    query = urllib.urlencode(query_dict, doseq=True)
    uri = (uri.scheme, uri.netloc, uri.path, uri.params, query, uri.fragment)#avoid magic number[4]
    
    #Construct a URL from a tuple as returned by urlparse()
    return urlparse.urlunparse(uri)
```

```python
#http://www.zillow.com/howto/api/GetDemographics.htm
def getDemographics_xml(zipcode):
    url_template_zillow = 'http://www.zillow.com/webservice/GetDemographics.htm?zws-id=&zip='
    url = set_url_parameter(url_template_zillow, 'zws-id', settings.zwsid)
    url = set_url_parameter(url, 'zip', zipcode)
    response = requests.get(url).content 

    return etree.fromstring(response)
```

```python
def parse_value(path, xml_doc):
    '''Take xml doc retrieved and parse certain value according to the path'''
#     value_list = []
    value = xml_doc.xpath(path)
    if value != []:
        return value[0].text
#         for item in value:            
#             value_list.append(item.text)
#     else:
#         value_list.append(0)
#     return value_list
```

```python
def getDemographicsByCity_xml(city, state):
    url_template_zillow = 'http://www.zillow.com/webservice/GetDemographics.htm?zws-id=&zip='
    url = set_url_parameter(url_template_zillow, 'zws-id', settings.zwsid)
    url = set_url_parameter(url, 'city', city)
    url = set_url_parameter(url, 'state', state)
    response = requests.get(url).content 

    return etree.fromstring(response)
```

```python
def getDemographicsByCity_df(city, state):
    Income_path = "//page[name='People']//attribute[name='Median Household Income']/values/city/value"
    doc = getDemographicsByCity_xml(city, state)
    income = parse_value(Income_path, doc)
    return income
getDemographicsByCity_xml("Oakland", "CA")
```

```python
def getDemographics_df(ziplist):
    Index_path = "//page[name='Affordability']//attribute[name='Zillow Home Value Index']/values/zip/value"
    SFH_path = "//page[name='Affordability']//attribute[name='Median Single Family Home Value']/values/zip/value"
    Condo_path = "//page[name='Affordability']//attribute[name='Median Condo Value']/values/zip/value"
    SalePrice_path = "//page[name='Affordability']//attribute[name='Median Sale Price']/values/zip/value"
    Persqft_path = "//page[name='Affordability']//attribute[name='Median Value Per Sq Ft']/values/zip/value"    
    One_year_path = "//page[name='Affordability']//attribute[name='1-Yr. Change']/values/zip/value"
    Income_path = "//page[name='People']//attribute[name='Median Household Income']/values/nation/value"
    
    demographics = {zipcode:[] for zipcode in ziplist}
    for zipcode in ziplist:
        doc = getDemographics_xml(zipcode)
#         parse_value(Income_path, doc)
        demographics[zipcode].append(parse_value(Income_path, doc))
    return DataFrame.from_dict(demographics, 'index')
# DataFrame([[zipcode, home_value, median_per_sqft] for zipcode, value in demographics for home_value, median_per_sqft in value])

getDemographics_df(oak_zip)
```

```python
#http://developer.trulia.com/docs/read/TruliaStats/getZipCodeStats
def getZipCodeStats_xml(zipcode,startDate,endDate):
    url_template_trulia = 'http://api.trulia.com/webservices.php?library=TruliaStats&function=getZipCodeStats&zipCode=&startDate=&endDate=&apikey='
    url = set_url_parameter(url_template_trulia, 'apikey', settings.apikey)
    url = set_url_parameter(url, 'zipCode', zipcode)
    url = set_url_parameter(url, 'startDate', startDate)
    url = set_url_parameter(url, 'endDate', endDate)
    response = requests.get(url).content   
    
    return etree.fromstring(response)
```

```python
def getDeepSearchResults_xml(addr, citystatezip):
    url_template_zillow = 'http://www.zillow.com/webservice/GetDeepSearchResults.htm?zws-id=&address=&citystatezip='
    url = set_url_parameter(url_template_zillow, 'zws-id', settings.zwsid)
    url = set_url_parameter(url, 'address', addr)
    url = set_url_parameter(url, 'citystatezip', citystatezip)
    response = requests.get(url).content
    
    return etree.fromstring(response)
#     return url

# getDeepSearchResults_xml("2114 Bigelow Ave", "Seattle, WA")
```

```python
#http://www.zillow.com/howto/api/GetDeepSearchResults.htm
def getDeepSearchResults(addr, citystatezip):
    '''Take in address and city+state combination and/or ZIP code, output '''
    
    lastSoldDate_path = "//response//lastSoldDate"
    lastSoldPrice_path = "//response//lastSoldPrice"
    latitude_path = "//response//address/latitude"
    longitude_path = "//response//address/longitude"
    
    doc = getDeepSearchResults_xml(addr, citystatezip)
    lastSoldDate = parse_value(lastSoldDate_path, doc)[0]
    lastSoldPrice = parse_value(lastSoldPrice_path, doc)[0]
    latitude = parse_value(latitude_path, doc)[0]
    longitude = parse_value(longitude_path, doc)[0]
    
    return Series([lastSoldDate, lastSoldPrice, latitude, longitude])

getDeepSearchResults("2114 Bigelow Ave", "Seattle, WA")    
```
