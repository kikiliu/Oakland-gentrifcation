### First, creating hash table, with key( tuple(city name,neigborhood name) )  -->   value(polygon)

```python
hash_zillow_polygons = {}

for record in fiona.open('ZillowNeighborhoods-CA.shp', 'r'):
    if record['properties']['CITY'].encode() == 'Oakland':
        polygon = shape(record['geometry'])
        neigborhood = record['properties']['NAME']
        #city = record['properties']['CITY']
        hash_zillow_polygons[neigborhood] = polygon
```

```python
print 'according to zillow, there are {0} neighborhoods in Oakland.'.format(len(hash_zillow_polygons))
```

### this code just shows how to use the dictionaries. The polygons are in fact correct, and I verified that no point can map to more than one polygon.

```python
point = Point(-122.1849, 37.73299) # a random point in Oakland

for p in hash_zillow_polygons:
    if hash_zillow_polygons[p].contains(point):
        print 'found point'
```

### Now, using the following method to create a neighborhood column in the dataframe.

```python
def create_neighborhood_column(row):
    point = Point(row['LONG'],row['LAT'])
    for hood in hash_zillow_polygons:
        polygon = hash_zillow_polygons.get(hood)
        if polygon.contains(point):
            return hood.__str__() # hood.encode() would work as well, but slower      
    return 'NOT FOUND' 
```

```python
# takes about 4 minutes to run
df['NEIGHBORHOOD'] = df.apply(lambda x: create_neighborhood_column(x), axis=1)
```

```python
df = df[df.NEIGHBORHOOD != 'NOT FOUND']
```

### I would really like to rid the dataframe of all crimes that have no real weight with regards to safety, or of little concern. One of te questions I was trying to answer, was: what would the same graphs per neighborood look like, if I was to do consolidate by crime-type. My guess was that Oakland would not be as scarry as it initially was- over 400,000 crimes in the dataframe initially. Lets try this:

#### These are the indicators I set as 1 for True, 0 for False within the dataframe: Violence,Property,Homicide,Assaults,Robbery,Shootings,Burglary,MV Theft,Rape,Weapons,Drugs,Sex It is important to note that a single crime can have multiple of these values checked at the same time.

#### Initially, I am making a function that marks crimes that are 'serious' if they have True values in either one of these: Violence, Homicide, Assaults, Shootings, Rape and Weapons. I chose to ignore car and drug related crimes, counting on the fact that if they were out of the scope of a 'normal' car theft or an arrest for drugs posession, they would probably be checked already with one of the more 'serious' marks. Same distinction with Property, Robbery, Burglery, and Sex indicators.

```python
def keep_serious_crimes(series):
    if series.Violence == 1:
        return 1
    if series.Homicide == 1:
        return 1
    if series.Assaults == 1:
        return 1
    if series.Shootings == 1:
        return 1
    if series.Rape == 1:
        return 1
    if series.Weapons == 1:
        return 1
    return 0
```

```python
df_crime_sort = df.copy()
```

```python
df_crime_sort['serious'] = df_crime_sort.apply(lambda x: keep_serious_crimes(x), axis=1)
```
