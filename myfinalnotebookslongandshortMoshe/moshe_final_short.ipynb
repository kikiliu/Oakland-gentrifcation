### I would really like to rid the dataframe of all crimes that have no real weight with regards to safety, or of little concern. ### These are the indicators I set as 1 for True, 0 for False within the dataframe: Violence,Property,Homicide,Assaults,Robbery,Shootings,Burglary,MV Theft,Rape,Weapons,Drugs,Sex It is important to note that a single crime can have multiple of these values checked at the same time.

### Initially, I am making a function that marks crimes that are 'serious' if they have True values in either one of these: Violence, Homicide, Assaults, Shootings, Rape and Weapons. I chose to ignore car and drug related crimes, counting on the fact that if they were out of the scope of a 'normal' car theft or an arrest for drugs posession, they would probably be checked already with one of the more 'serious' marks. Same distinction with Property, Robbery, Burglery, and Sex indicators.

```python
def keep_serious_crimes(series):
    if series.Violence == 1:
        return 1
    if series.Homicide == 1:
        return 1
    if series.Assaults == 1:
        return 1
    if series.Shootings == 1:
        return 1
    if series.Rape == 1:
        return 1
    if series.Weapons == 1:
        return 1
    return 0
```

```python
df_crime_sort = df.copy()
df_crime_sort = df_crime_sort[['NEIGHBORHOOD','YEAROCC','Violence','Homicide','Assaults','Shootings','Rape','Weapons']]
df_crime_sort['serious'] = df_crime_sort.apply(lambda x: keep_serious_crimes(x), axis=1)
df_crime_sort = df_crime_sort[['NEIGHBORHOOD','YEAROCC','serious']]
```

```python
len(df_crime_sort[df_crime_sort.serious == 1])
```

```python
df_drop_non_serious = df_crime_sort.copy()
```

```python
df_drop_non_serious = df_drop_non_serious[df_drop_non_serious.serious == 1]
```

```python
hood_group_serious = df_drop_non_serious.groupby('NEIGHBORHOOD').count()
```

```python
hood_group_serious_sorted = hood_group_serious.sort(column='YEAROCC',ascending=False)
```
